```python
#import instructions
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# Read the CSV file (dataset)
train = pd.read_csv("emnist-letters-train.csv", header=None)
test = pd.read_csv("emnist-letters-test.csv" , header=None)
```

```python
train.head()
```

```python
# # Splite datasets

# training_letters
X_train1 = np.array(train.iloc[:,1:].values)
y_train1 = np.array(train.iloc[:,0].values)

# testing_labels
X_test2 = np.array(test.iloc[:,1:].values)
y_test2 = np.array(test.iloc[:,0].values)

print(X_train1.shape)
print(y_train1.shape)

print(X_test2.shape)
print(y_test2.shape)
```

```python
X_train1.shape
```

```python
#Reading the file
label_map = pd.read_csv('New Text Document.txt', delimiter = ' ', index_col=0,header=None, squeeze=True)
label_dict = {}
#Running a loop for ASCII equivalent to character conversion
for index, label in enumerate(label_map):
    label_dict[index] = chr(label)
```

```python
# label_dict.pop(-1)
label_dict
```

```python
# Before normalization
print("a sample of the data Before normalization : \n", X_test2[222])


print("#"*100)
# Random samples visualization
plt.figure(figsize = (12,10))
row, colums = 5, 5
for i in range(25):  
    plt.subplot(colums, row, i+1)
    plt.imshow(X_train1[i].reshape([28,28]),interpolation='nearest', cmap='Greys')
plt.show()
```

```python

# Normalise and reshape data
train_images = X_train1 / 255.0
test_images = X_test2 / 255.0

# train_images_size = 28 * 28
# train_images = train_images.reshape(train_images.shape[0], 28, 28, 1)
# test_images_size = 28 * 28
# test_images = test_images.reshape(test_images.shape[0], 28, 28, 1)
```

```python
# After normalization
print("a sample of the data After normalization : ", test_images[222])

print("#"*50)
```

```python
# Transform labels
number_of_classes = 28

y1 = tf.keras.utils.to_categorical(y_train1, number_of_classes)
y2 = tf.keras.utils.to_categorical(y_test2, number_of_classes)
```

```python
y1
```

```python
# Display the shape of the training and testing data
print("X_train shape : ", train_images.shape)
print("y_train shape : ", test_images.shape)
print("y_train shape : ", y1.shape)
```

```python
# Split 10% validation 
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train_images, 
                                                  y1, 
                                                  test_size= 0.1, 
                                                  random_state=88)
```

# ANN(Artificial Neural Network)
 ![](https://thumbs.gfycat.com/AdorableJoyfulLemming-max-1mb.gif)

```python
from keras.models import Sequential, load_model
from keras.layers import Dense
model = Sequential()


model.add(Dense(784, input_shape=(784,), kernel_initializer='normal', activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(28, kernel_initializer='normal', activation='softmax'))
```

 # EarlyStopping

 ![](https://miro.medium.com/max/474/1*wZg_RQHPRtn62dDp2Ez86A.jpeg)

```python
# Optimizing Algorithm for Backpropagation
from keras.callbacks import EarlyStopping, ModelCheckpoint
optimizer_name = 'adam'

model.compile(loss='categorical_crossentropy', optimizer=optimizer_name, metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, mode='min')
mcp_save = ModelCheckpoint('my_model3.h5', save_best_only=True, monitor='val_loss', verbose=1, mode='auto')
```

```python
#Model Fitting

history = model.fit(x_train,
                    y_train, 
                    epochs=30, 
                    batch_size=50, 
                    verbose=1, 
                    validation_split=0.1,
                    callbacks=[early_stopping, mcp_save])
```

```python
accuracy=model.evaluate(x= x_val,y=y_val,batch_size=32)
print(accuracy[1])
```

```python
def plotgraph(epochs, acc, val_acc):
    # Plot training & validation accuracy values
    plt.plot(epochs, acc, 'b')
    plt.plot(epochs, val_acc, 'r')
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Val'], loc='upper left')
    plt.show()
    
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(acc)+1)
```

```python
plotgraph(epochs, acc, val_acc)
```

```python
# loss curve
print('Loss Curve')
plotgraph(epochs, loss, val_loss)
```

```python
def plot_roc_curve(fpr, tpr):
    plt.figure(figsize=(10,6))
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
     
```

```python
plot_roc_curve(fpr_keras, tpr_keras)
```
